## EJERCICIO 4

Imprimir los números de 1 hasta un número natural n dado, cada uno con su respectivo factorial

```python
if __name__ == "__main__":
  n = int(input("Ingrese un número natural:")) # Ingresar el numero
  fac = 1
  for i in range (1, n + 1): # Seleccionar el rango y continuar con el siguiente numero
    fac *= i
  print("El factorial de", n , "es:", fac ,) # Imprimir el resultado
```

## EJERCICIO 8

Diseñar una función que permita calcular una aproximación de la función exponencial alrededor de 0 para cualquier valor x (real), utilizando los primeros n términos de la serie de Taylor. Nota: use math para traer la función exponencial y mostrar la diferencia entre el valor real y la aproximación.

e x ≈ e x p ( x , n ) ≈ ∑ i = 0 n x i i !

```python
import math

def exp_aprox(x, n):
    suma = 0  # Inicializa la suma
    for i in range(n + 1):  # Rango desde 0 hasta n
        t = (x ** i) / math.factorial(i)  # Calcula el termino con la función exponencial
        suma += t # Suma el término a la secuencia
    return suma

if __name__ == "__main__":
    x = float(input("Ingrese el valor de x: "))  # Ingresar x y n
    n = int(input("Ingrese el número de términos de la serie: "))
    aprox = exp_aprox(x, n) # Variables de aproximación y el valor real
    real = math.exp(x)

    print("Aproximación de e^x usando", n, "términos:", aprox) # Imprimir los resultados
    print("Valor real de e^x:", real)
    print("Diferencia entre el valor real y la aproximación:", abs(real - aprox))
```

## EJERCICIO 9

Diseñar una función que permita calcular una aproximación de la función seno alrededor de 0 para cualquier valor x (real), utilizando los primeros n términos de la serie de Maclaurin. Nota: use math para traer la función seno y mostrar la diferencia entre el valor real y la aproximación.

s i n ( x ) ≈ s i n ( x , n ) ≈ ∑ i = 0 n ( − 1 ) i x 2 i + 1 ( 2 i + 1 ) !



```python
import math

def sin_aprox(x, n):
    suma = 0  # Inicializa la suma
    for i in range(n + 1):  # Rango desde 0 hasta n
        termino = ((-1) ** i) * (x ** (2 * i + 1)) / math.factorial(2 * i + 1)  # Calcula el término de la formula
        suma += termino  # Suma el término a la secuencia
    return suma

if __name__ == "__main__":
    x = float(input("Ingrese el valor de x (en radianes): ")) # Ingresar x y n
    n = int(input("Ingrese el número de términos de la serie: "))
    aprox = sin_aprox(x, n)  # Calcular la aproximación y el valor real
    real = math.sin(x)

    print("Aproximación de sin(x) usando", n, "términos:", aprox)  # Imprimir los resultados
    print("Valor real de sin(x):", real)
    print("Diferencia entre el valor real y la aproximación:", abs(real - aprox))
```
